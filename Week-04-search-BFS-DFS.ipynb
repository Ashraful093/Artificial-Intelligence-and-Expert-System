# BFS

```python
def bfs(startingNode, destinationNode):
    reached = [] # to store reached (generated) nodes
    expand_sequence = [] # to store the order of nodes expanded
    frontier = []

    snode = {'City':startingNode,'Path cost':0, 'Path':[startingNode]}
    frontier.append(snode)    # insert the starting node
    reached.append(startingNode)
    
    while len(frontier) > 0:
        print(frontier)
        unode = frontier.pop(0)     # selceting the node to expand
        u = unode['City']
        expand_sequence.append(u)
        print(u+" expanding")
        # expand u

        for v in romanian_map[u].keys():
            if v not in expand_sequence: # process if v is not reached (generated) yet
                print(v+" Generated")                
                cost = unode['Path cost'] + romanian_map[u][v] 
                path =  unode['Path']+ [v]
                vnode = {'City': v,'Path cost': cost,'Path': path}
                frontier.append(vnode)# store the new child node in the frontier
                reached.append(v)
                # goal test
                if v == destinationNode:
                    #print(vnode)
                    print(expand_sequence)
                    return vnode # breaking from while
    print('Failed')   
```

```python
bfs('Arad', 'Bucharest') # change the soure and destination
```

**Uniform Cost Search**

```python
from queue import PriorityQueue


def ucs(snode, dnode):
    expanded = [] # to store expanded nodes
    frontier = PriorityQueue()

    node = (0,snode,[snode]) #(g(n) = path cost, city name, path)
    frontier.put(node)
    
    while not frontier.empty():
        node = frontier.get() # select a node for expansion
        u = node[1] 
        print()
        print(u+ " selected. Cost: "+ str(node[0]))
        if u == dnode:# goal test            
            print("*** GOAL TEST SATISFIED ***")
            print()
            return node # path
        expanded.append(u)
        print(u+" expanding...")
        for v in rmap2[u].keys():#expanding            
            if v not in expanded:
                g = node[0] + rmap2[u][v]
                path = node[2] + [v]
                frontier.put((g,v,path))
                print(u + "->"+v + ", Path: "+ str(path) + ", Cost:"+ str(g))
        print(frontier.queue)
    print('Failed')
```
